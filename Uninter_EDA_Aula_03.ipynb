# Bibliotecas e Importação dos dados

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dayanebravo/AED/main/pes_2012.csv')
df.head() 
```

```python
ndf = df.dropna() 
```

```python
# Converte para numérico, transforma em float e substitui os erros por NaN
df['V4718'] = pd.to_numeric(df['V4718'], errors='coerce')
df['V4720'] = pd.to_numeric(df['V4720'], errors='coerce')   
```

## Bibliotecas para gráficos

```python
import matplotlib.pyplot as plt
import seaborn as sn
sn.set(style='ticks', palette='light:m_r') #estilo e cor dos gráficos
```

## Regra para contagem de classes

### Regra do logaritmo: image.png

```python
sizes= df.V8005 #Coluna analisada
plt.figure(figsize=(8,8))

# Calculo da quantidade de classes
k = int(1 + 3.3 * np.log(len(sizes)))
print(k)

sn.histplot(sizes, bins=k) # quantidade de classes/intervalos - largura calculada automaticamente
plt.ylabel('qtd pessoas')
plt.xlabel('idade')
plt.show()
```

### Regra da potência

image.png

```python
sizes= df.V8005 #Coluna analisada
plt.figure(figsize=(8,8))

# Calculo da quantidade de classes
import math
# lembrando que 2^k=n é o mesmo que log_2(n)=k
c = math.log(len(sizes), 2) # fórmula do log na base 2
k = int(c+1) # k precisa ser inteiro e soma 1 para garantir que será maior que o valor de n
print(k)

sn.histplot(sizes, bins=k) # quantidade de classes/intervalos - largura calculada automaticamente
plt.ylabel('qtd pessoas')
plt.xlabel('idade')
plt.show()
```

## Histograma

```python
sizes= df.V8005 #Coluna analisada

sn.histplot(sizes, bins=15, cumulative=True)
plt.ylabel('qtd pessoas')
plt.xlabel('idade')
plt.show()
```

## Distribuição

```python
sizes= df.V8005 #Coluna analisada

sn.displot(sizes, bins=15, cumulative=False)

plt.ylabel('qtd pessoas')
plt.xlabel('idade')
plt.show()
```

```python
sizes= df.V8005 #Coluna analisada

sn.distplot(sizes) # esse gráfico será descontinuado em algum momento

plt.ylabel('qtd pessoas')
plt.xlabel('idade')
plt.show()
```

## Densidade

```python
sizes= df.V8005 #Coluna analisada

sn.kdeplot(data=sizes, fill=True) # tipo do gráfico com preenchimento de cor

plt.ylabel('qtd pessoas')
plt.xlabel('idade')
plt.show()
```

## Poligono de frequência

```python
sizes= df.V0404.value_counts() #Coluna analisada

sn.relplot(data=sizes, kind="line", height=5) # tipo do gráfico e altura

plt.ylabel('qtd pessoas')
plt.xlabel('idade')
plt.show()
```

## Gráfico de dispersão

```python
plt.figure(figsize=(8,8))

sizes= df.V0404.value_counts() #Coluna analisada

# sn.relplot(x=df.V8005, y=df.V4720, kind="scatter") # uma maneira de fazer o gráfico de dispersão
sn.scatterplot(x=df.V8005, y=df.V4720) # outra maneira de fazer o gráfico de dispersão

plt.ylabel('renda')
plt.xlabel('idade')
plt.show()
```
